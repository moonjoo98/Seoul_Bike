```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

## 정류장 마스터 정보

```python
master = pd.read_csv('public_answer_data/서울시 따릉이대여소 마스터 정보.csv', encoding='CP949')
master
```

```python
# 자치구 도출
master['자치구'] = master['주소1'].apply(lambda x: x.split()[1] if '구' in x else 'no')
```

```python
master
```

```python
# 필요한 자치구의 정류장 정보만 저장
need_master = master[master['자치구'].isin(['광진구', '동대문구', '성동구', '중랑구'])].reset_index(drop=True)
need_master
```

```python
need_id = list(need_master['대여소_ID'])
```

## 20220101~20220630 따릉이 대여 데이터

```python
month = []
for i in range(1, 7):
    month.append('0'+str(i))
```

```python
month
```

```python
day = []
for i in range(1, 32):
    if i < 10:
        day.append('0'+str(i))
    else:
        day.append(str(i))
```

```python
print(day)
```

```python
f = pd.DataFrame(columns = ['기준_날짜', '기준_시간대', '시작_대여소_ID', '종료_대여소_ID', '전체건수', '전체_이용_분', '전체_이용_거리'])

for m in tqdm(month):
    for d in day:
        
        try:
            df = pd.read_csv(f'public_answer_data/tpss_bcycl_od_statnhm_2022{m}/tpss_bcycl_od_statnhm_2022{m}{d}.csv', encoding='CP949')
            df.columns = ['기준_날짜', '기준_시간대', '시작_대여소_ID', '종료_대여소_ID', '전체건수', '전체_이용_분', '전체_이용_거리']
            f = pd.concat([f, df], axis=0)
            
        except UnicodeDecodeError:
            df = pd.read_csv(f'public_answer_data/tpss_bcycl_od_statnhm_2022{m}/tpss_bcycl_od_statnhm_2022{m}{d}.csv')
            df.columns = ['기준_날짜', '기준_시간대', '시작_대여소_ID', '종료_대여소_ID', '전체건수', '전체_이용_분', '전체_이용_거리']
            f = pd.concat([f, df], axis=0)
        
        except FileNotFoundError:
            continue
```

```python
f.shape
```

```python
f.isna().sum()
```

```python
# 마스터 정보 결합
total = pd.merge(f, need_master, how='left', left_on='시작_대여소_ID', right_on='대여소_ID')
total
```

```python
# 필요한 자치구의 정류장 정보만 저장
need = total[total['시작_대여소_ID'].isin(need_id)]
need
```

```python
summary = need.groupby(['자치구', '기준_날짜'])[['전체건수']].sum().reset_index()
summary
```

```python
public_answer = pd.pivot_table(summary,                # 피벗할 데이터프레임
                               index = '기준_날짜',    # 행 위치에 들어갈 열
                               columns = '자치구',    # 열 위치에 들어갈 열
                               values = '전체건수')     # 데이터로 사용할 열 
```

```python
answer = public_answer.reset_index()
answer
```

```python
answer.to_csv('self_evaluation/public_answer_20220101_20220630.csv', index=False)
```
