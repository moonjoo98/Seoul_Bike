```python
def change_age(x):
    if x == '~10대':
        return 'AGE_001'
    elif x == "20대":
        return 'AGE_002'
    elif x == "30대":
        return 'AGE_003'
    elif x == "40대":
        return "AGE_004"
    elif x == "50대":
        return "AGE_005"
    elif x == "60대":
        return "AGE_006"
    elif x== "70대~":
        return "AGE_007"
    elif x in ['nan', '\\N']:
        return "AGE_008"
    else:
        return x
```
