```python
def extract_gu(x):
    try:
        gu = str(x).split(' ')[1]
        return gu
    except:
        gu = 'nan'
        return gu
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dacon/ddareung/ex_data/21년.csv')

df2['ARS-ID'] = df2['ARS-ID'].apply(lambda x : '0'+str(x) if len(str(x))==4 else str(x))
df2['구'] = df2['_PARCEL_AD'].apply(lambda x : extract_gu(x))

lst = ['광진구', '동대문구', '성동구', '중랑구']
df2 = df2.query("구 in @lst").reset_index(drop=True)

df2.head()
```

```python
df2.shape
```

```python
df3 = df.merge(df2[['ARS-ID', '구']], left_on='버스정류장ARS번호', right_on='ARS-ID')
```

```python
df3
```

```python
df3['구'].unique()
```

```python
df4 = df3.groupby(['사용일자', '구'])['하차총승객수'].mean().reset_index()
df5 = df3.groupby(['사용일자', '구'])['하차총승객수'].sum().reset_index()
df6 = df3.groupby(['사용일자', '구'])['하차총승객수'].size().reset_index()
df7 = df4.merge(df5, on=['사용일자', '구'])
df7 = df7.merge(df6, on=['사용일자', '구'])
df7.columns = ['사용일자', '구', '하차총승객수_mean', '하차총승객수_sum', '하차총승객수_size']
df7
```
