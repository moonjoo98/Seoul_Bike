```python
lst = list(set(b) - set(a))
```

```python
hacha = hacha.values.reshape(1457, -1)
```

```python
for i in range(data.shape[0]):
    for j in range(hacha.shape[0]):
        if data['일시'][i] == hacha[j][0]:
            data.iloc[i, 9:12] = hacha[j][2:5]
            data.iloc[i, 12:15] = hacha[j][7:10]
            data.iloc[i, 15:18] = hacha[j][12:15]
            data.iloc[i, 18:] = hacha[j][-3:]
        else:
            continue
```
